Exception Handling -> try, except

```python
try:
  n=int(input('Enter a number : '))
except:
  n=0
  print('invalid number entered...')

sq=n*n
print(f"square of {n} is {sq}")
```

```python
custnm=input('Enter customer name : ')
try:
  purchamt=float(input('Enter purchase amount : '))
except ValueError:
  purchamt=0
  print('invalid purchase amount...reset to zero')
finally:
  print('Thanks for shopping with us!')

discount=purchamt*10/100
bill=purchamt-discount
print(f"{custnm} your discount is {discount} and net bill is {bill}")
```

**Different except for different errors**

```python
try:
  x=int(input('Enter first number : '))
  y=int(input('Enter second number : '))
  res=x/y
except ValueError:
  res=0
  print('invalid input')
except ZeroDivisionError:
  res=x
  print('cant divide by zero..divided by 1 instead')

print(f"result of division is {res}")
```

**Common except for all errors**

```python
try:
  x=int(input('Enter first number : '))
  y=int(input('Enter second number : '))
  res=x/y
except:
  res=0
  print('invalid input or divide by zero error')

print(f"result of division is {res}")
```

```python
i=int(input('Enter index : '))
try:
  print(lst[i])
except:
  print('not found')
```

```python
try:
  i=int(input('Enter index : '))
  try:
    print(lst[i])
  except IndexError:
    print('not found')
except ValueError:
  print("index numbers can only be integers")
```
