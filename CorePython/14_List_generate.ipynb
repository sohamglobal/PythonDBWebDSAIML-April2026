```python
lst=[9,13,45,26,1,30,10,42,15,79]
sq=[]
for n in lst:
  sq.append(n*n)

print(sq)
```

```python
lst=[9,13,45,26,1,30,10,42,15,79]
sq=[n*n for n in lst]
print(sq)
```

```python
lst=['praffull','sharayu','soham','shailaja','buttler']
u=[nm.upper() for nm in lst]
print(u)
```

```python
lst=['praffull','sharayu','soham','shailaja','buttler']
res=[nm for nm in lst if nm.startswith('sh')]
print(res)
```

```python
lst=[9,13,45,26,1,30,10,42,15,79]
odd=[n for n in lst if n%2==1]
print(odd)
```

```python
lst=[9,13,45,26,1,30,10,42,15,79]
res=[n for n in lst if n%3==0]
print(res)
```

```python
l1=[2,14,25,57,13]
l2=[1,5,6,13,8,9,26,14]
res=[x for x in l1 if x in l2]
print(res)
```

```python
l1=[2,14,25,57,13]
l2=[1,5,6,13,8,9,26,14]
res=[x for x in l1 if x not in l2]
print(res)
```
