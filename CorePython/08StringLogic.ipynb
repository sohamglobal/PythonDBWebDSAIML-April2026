**Memory utilization in Strings**

```python
a="sohamglobal"
b="sohamglobal"
print(id(a)==id(b))

x="praffull"
y="Praffull"
print(id(x)==id(y))
```

```python
nm="Microsoft Corporation"
print(nm[0])
print(nm[11])
print(nm[3:15])
print(nm[-1])
print(nm[-5])
```

```python
nm="sohamglobal"
print(nm[::-1])
```

```python
s="madam"
s1=s[::-1]

if s==s1:
  print("it is a palindrome")
else:
  print("it is not a palindrome")
```

```python
s="technology Is power"
vowels="aeiouAEIOU"
count=0

for ch in s:
  if ch in vowels:
    count+=1

print(f"total vowels in the string are {count}")
```

```python
s1="night"
s2="thing"
print(sorted(s1))
print(sorted(s2))

if sorted(s1)==sorted(s2):
  print("anagram")
else:
  print("not anagram")
```
