```python
class Player:
  def __init__(self):
    self.name="Not available"
    self.country="NA"
  def takedata(self):
    self.name=input('Enter player name : ')
    self.country=input("Enter country : ")
  def showdata(self):
    print(f"Player name is {self.name}")
    print(f"Country is {self.country}")
```
