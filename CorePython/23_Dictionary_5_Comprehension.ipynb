**Dictionary**

```python
sqr={x:x**2 for x in range(1,11)}
print(sqr)
```

```python
sqodd={x:x**2 for x in range(51,76) if x%2==1}
print(sqodd)
```

```python
eqres={x:x*x+13*x+9 for x in range(1,11)}
print(eqres)
```

```python
prices={'apple':50,'banana':30,'mango':120,'orange':40}
discounted={item:price-(price*13/100) for item,price in prices.items()}
print(discounted)
```
