**Arithmetic operators**

```python
x=9
y=13
print(x+y)
print(x-y)
print(x*y)
print(x/y)
#floor division
print(y//x)
#modulus - remainder
print(y%x)
print(23%7)
print(6**2)
print(2**3)
```

**Relational Operator**

```python
x=5
print(x==4)
password="Chelsea"
print(password=="chelsea")

print(10>15)
print(10<15)
print(x<=5)
print(x>=4)
print(x!=10)
```

**Assignment Operators**

```python
x=9
x+=3
print(x)
x-=4
x*=2
x/=3
x//=2
x%=4
print(x)
```

**Logical Operators**

```python
print(10>5)
print(7<3)
print(10>5 and 7<3)
print(10>5 or 7<3)
print(not 10>5)
```

**Membership operators**

```python
print('e' in 'praffull')
print('s' not in 'praffull')
skills=['java','python','spring boot','jpa','openai','sql']
print('mongodb' in skills)
```

**Ternary operator**

```python
marks=27
result='pass' if marks>=35 else 'fail'
print(result)
```
