```python
nm=input('Enter customer name : ')
amt=float(input('Enter purchase amount : '))

if amt>=20000:
  print("13% discount applied")
  disc=amt*13/100
else:
  print("9% discount applied")
  disc=amt*9/100
print(f'Dear {nm}, you will get discount {disc} on your shopping of {amt}')
```
