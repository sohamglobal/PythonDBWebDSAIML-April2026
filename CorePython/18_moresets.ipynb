```python
lst=[5,1,2,1,7,2,9,1,3,6,1,9,7,2,4,2]
lst=list(set(lst))
print(lst)
```

**membership testing**

```python
auth_users={'sharayu','praffull','shailaja','megha'}
user=input('enter userid : ')

if user in auth_users:
  print('access granted')
else:
  print('access denied')
```

**set ops on list**

```python
l1=['java','spring boot','jpa','python','sql']
l2=['python','pandas','django','flask','sql','java']
```

```python
res=list(set(l1).intersection(l2))
print(res)
```
