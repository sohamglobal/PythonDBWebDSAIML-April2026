**ByteArray**

bytes is immutable whereas bytearray is mutable

```python
ba1=bytearray(b'Praffull')
print(ba1)
ba1[0]=83
print(ba1)
```

```python
ba1.append(63)
print(ba1)
```

```python
ba1.insert(0,35)
print(ba1)
```

```python
ba1.remove(97)
print(ba1)
```
