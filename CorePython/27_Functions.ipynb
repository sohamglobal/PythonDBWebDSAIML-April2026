```python
def calcdiscount(amount):
  if amount>20000:
    discount=int(amount*13/100)
  else:
    discount=int(amount*9/100)

  print(f"Amount : {amount} and Discount : {discount}")
```

```python
calcdiscount(7230)
calcdiscount(21580)
```

**Variable positional arguments**

```python
def congrats(*names):
  #print(f"congrats {names}")
  for nm in names:
    print(f"congrats {nm}")
```

```python
congrats('shailaja')
congrats('praffull','sharayu','soham','megha')
```

```python
def summarize(*args):
  print(f"number of values {len(args)}")
  print(f"sum of values is {sum(args)}")
  print(f"largest value is {max(args)}")
  print(f"smalles value is {min(args)}")
```

```python
summarize(9,13,26,45,1)
```

```python
summarize(12,65)
```
