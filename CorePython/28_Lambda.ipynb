**Lambda function is a small, single line, anonymous function**

```python
def calcsquare(x):
  sq=x**2
  return sq
```

```python
print("square is ",calcsquare(9))
print(f"Square is {calcsquare(13)}")
s=calcsquare(26)
print(f"square is {s}")
```

lambda parameters:return value

```python
sq=lambda x:x**2
```

```python
print('square is',sq(5))
print('square is',sq(12))
```

```python
sum=lambda a,b:a+b
```

```python
print(sum(34,56))
```

```python
d=lambda amount:amount*10/100
```
