```python
a={1,2,3,4,5}
b={6,7,5,8,4}
#union - combine all values from both the sets
res=a.union(b)
print(res)

#intersection - common values from both sets
res=a.intersection(b)
print(res)

#difference - values in the first set but not in second
res=a.difference(b)
print(res)

#symmetric difference - all values from both sets except the common
res=a.symmetric_difference(b)
print(res)
```

```python
microsoft={'Windows','Office','Cloud','XBox','LinkedIn','GitHub','Teams','Search','Browser'}
google={'Search','Gmail','Drive','Cloud','Maps','Android','Browser'}
apple={'iPhone','Mac','iPad','Cloud','Browser','Music','Watch'}
```

**Intersection &**

```python
common_services=microsoft & google & apple
print(common_services)
```

**Union |**

```python
all_services=microsoft | google | apple
print(all_services)
```

**DIfference -**

```python
microsoft_exclusive=microsoft - google - apple
print(microsoft_exclusive)
```

**Symmetric difference ^**

```python
only_exclusive=microsoft^google
print(only_exclusive)
```
