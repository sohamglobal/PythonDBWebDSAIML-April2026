```python
tech={
  "os":"windows 11 pro",
  "language":"java 21 lts",
  "database":"oracle"
}
print(tech)
```

**Add new data to the dictionary**

```python
tech['cloud']='microsoft azure'
print(tech)
```

**Modify existing data**

```python
tech['database']='mongodb atlas'
print(tech)
```

A dictionary can't have duplicate keys; values are overwritten

```python
tech.update({'framework':'spring boot 3','aiplatform':'deepseek'})
print(tech)
```

```python
tech['experience']=7
print(tech)
```

```python
tech['experience']+=1
print(tech)
```

**Remove an item**

```python
del tech['os']
print(tech)
```

```python
f=tech.pop('framework')
print(f)
print(tech)
```

```python
tech.clear()
print(tech)
```
